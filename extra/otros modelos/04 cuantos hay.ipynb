genial pregunta, Heri. “No hay hecho bruto: toda observación es ya teoría” (Einstein). Lo mismo con la **volatilidad**: cada especificación ARCH es una ontología distinta de cómo la información se sedimenta en σ²ₜ. Abajo tienes un **mapa canónico** (los más usados en la literatura). Para cada modelo: **año**, **ecuación** (forma mínima) y **función**.

# Univariados básicos (simétricos)

* **ARCH(q)** — *1982*
  εₜ = σₜ zₜ, zₜ∼i.i.d.(0,1); σ²ₜ = ω + Σᵢ₌₁^q αᵢ ε²ₜ₋ᵢ.
  Captura *clustering* de varianza con memoria corta (colas pesadas vía zₜ no-Gauss). ([econ.uiuc.edu][1], [JSTOR][2])

* **GARCH(p,q)** — *1986*
  σ²ₜ = ω + Σᵢ αᵢ ε²ₜ₋ᵢ + Σⱼ βⱼ σ²ₜ₋ⱼ.
  Parca en parámetros, memoria geométrica; estándar de facto. ([public.econ.duke.edu][3], [ScienceDirect][4])

* **IGARCH** — *1986*
  Igual que GARCH con Σαᵢ + Σβⱼ = 1 ⇒ persistencia “integrada” (sin varianza incondicional finita). Útil cuando el choque afecta “para siempre”. ([ScienceDirect][5], [stat.tugraz.at][6])

* **GARCH-M** (*in-Mean*) — *1987*
  yₜ = μ + λ·g(σ²ₜ) + εₜ (típicamente g(x)=x o √x) con σ²ₜ GARCH.
  Permite premio por riesgo dependiente de la volatilidad. ([IDEAS/RePEc][7])

# Univariados con asimetría (leverage)

* **EGARCH(p,q)** — *1991*
  log σ²ₜ = ω + Σⱼ βⱼ log σ²ₜ₋ⱼ + Σᵢ αᵢ(|zₜ₋ᵢ|−E|z|) + Σᵢ γᵢ zₜ₋ᵢ.
  Asimetría vía γ; sin restricciones de positividad (porque modela log-varianza). ([JSTOR][8], [finance.martinsewell.com][9])

* **GJR-GARCH (Threshold GARCH)** — *1993*
  σ²ₜ = ω + α ε²ₜ₋₁ + γ·I{εₜ₋₁<0}·ε²ₜ₋₁ + β σ²ₜ₋₁.
  Choques negativos pesan más (leverage). ([faculty.washington.edu][10])

* **TARCH / Zakoian** — *1994*
  σₜ = ω + α|εₜ₋₁| + γ·εₜ₋₁⁻ + β σₜ₋₁ (o variantes en σ²).
  Asimetría en la **desviación** condicional (no sólo varianza). ([ScienceDirect][11])

* **APARCH** — *1993*
  σᵟₜ = ω + Σᵢ αᵢ(|εₜ₋ᵢ| − γᵢ εₜ₋ᵢ)ᵟ + Σⱼ βⱼ σᵟₜ₋ⱼ, δ>0.
  Un “meta-modelo” que anida GARCH/TARCH/GJR según δ,γ. ([ScienceDirect][12], [ResearchGate][13])

* **NAGARCH / NGARCH (Engle–Ng)** — *1993*
  σ²ₜ = ω + α(εₜ₋₁ − θ σₜ₋₁)² + β σ²ₜ₋₁.
  Asimetría mediante desplazamiento del “news impact curve”. ([NBER][14], [finance.martinsewell.com][15])

* **QARCH (Quadratic ARCH)** — *1995*
  σ²ₜ = ω + a′xₜ₋₁ + x′ₜ₋₁ B xₜ₋₁ (x contiene innovaciones rezagadas); permite respuestas cuadráticas y asimétricas muy generales. ([Oxford Academic][16], [econpapers.repec.org][17])

# Memoria larga / hiper-memoria

* **FIGARCH(p,d,q)** — *1996*
  (1 − β(L))σ²ₜ = ω + {1 − β(L) − [1 − φ(L)](1−L)^d} ε²ₜ, 0\<d<1.
  Decaimiento hiperbólico de la memoria en varianza. ([ScienceDirect][18], [econpapers.repec.org][19])

* **FIEGARCH** — *1996*
  Variante fraccional de EGARCH: (1−β(L))(log σ²ₜ − ω) = \[1 − (1−L)^d]·g(zₜ₋₁).
  Captura asimetría **y** memoria larga. ([IDEAS/RePEc][20], [econ.duke.edu][21])

* **HYGARCH (Hyperbolic GARCH)** — *2004*
  Sustituye el operador fraccional por pesos hiperbólicos flexibles; intermedia entre IGARCH y FIGARCH. Útil en transiciones de régimen suaves. ([Tandfonline][22], [IDEAS/RePEc][23])

# Estructura por componentes / medidas realizadas

* **Component-GARCH (CGARCH)** — *1999*
  σ²ₜ = qₜ + α(ε²ₜ₋₁ − qₜ₋₁) + β(σ²ₜ₋₁ − qₜ₋₁), qₜ = ω + ρ qₜ₋₁.
  Descompone volatilidad en **tendencia** (largo plazo) y **ciclo**. ([Stern School of Business][24], [SCIRP][25])

* **Spline-GARCH** — *2008*
  σ²ₜ = GARCH(“unit-root”) × exp(spline lento); enlaza macro de baja frecuencia con shocks de alta frecuencia. ([Oxford Academic][26], [SSRN][27])

* **GARCH-MIDAS** — *2013*
  σ²ₜ = τₜ(θ; variables macro de baja frecuencia) × gₜ (GARCH diario).
  Separa **largo plazo** macro (MIDAS) y **corto plazo** financiero. ([ScienceDirect][28])

* **Realized-GARCH** — *2012*
  rₜ = σₜ zₜ; log σ²ₜ = ω + β log σ²ₜ₋₁ + γ log xₜ₋₁ + …; log xₜ = ξ + φ log σ²ₜ + τ zₜ + uₜ, donde xₜ es una medida realizada (RV). Mejora filtrado/forecast usando alta frecuencia. ([Wiley Online Library][29], [JSTOR][30])

# Multivariantes (covarianzas dinámicas)

* **VECH / VEC-GARCH** — *1988*
  vech(Hₜ) = c + A·vech(εₜ₋₁ε′ₜ₋₁) + B·vech(Hₜ₋₁). General, pero con muchos parámetros. ([swopec.hhs.se][31])

* **CCC-GARCH (Constant Conditional Correlation)** — *1990*
  Hₜ = Dₜ R Dₜ (R constante, Dₜ = diag(σₜ)); reduce dimensionalidad suponiendo correlación fija. ([public.econ.duke.edu][32])

* **BEKK-GARCH** — *1995*
  Hₜ = C′C + A′εₜ₋₁ε′ₜ₋₁A + B′Hₜ₋₁B. Parametrización que garantiza Hₜ≻0 y es invariante. ([kroner.com][33], [econpapers.repec.org][34])

* **DCC-GARCH** — *2002*
  Qₜ = (1−a−b)S + a uₜ₋₁u′ₜ₋₁ + b Qₜ₋₁; Rₜ = diag(Qₜ)^{-½} Qₜ diag(Qₜ)^{-½}. Correlaciones **time-varying** con paso en dos etapas. ([JSTOR][35], [econpapers.repec.org][36])

# Modelos para *pricing* (discretos afines)

* **Heston–Nandi GARCH** — *2000*
  Especificación GARCH afín bajo **Q** que entrega **precio cerrado** para calls europeas; puente discreto con Heston (1993). Crucial para ligar dinámica histórica y valuación. ([JSTOR][37], [IDEAS/RePEc][38])

# Otras extensiones notables

* **NARCH (Higgins–Bera, 1992)**: no linealidad en |ε|^δ (flexibiliza respuesta a choques). ([JSTOR][39])
* **GARCH-X**: incluye regresores exógenos en σ²ₜ (política monetaria, liquidez, etc.); familia práctica más que un único paper seminal. (véase panoramas en Engle/Bollerslev/Teräsvirta). ([EconStor][40])
* **HARCH (Müller–Dacorogna et al., \~1997)**: σ²ₜ como suma de “parciales” a múltiples horizontes (1d, 5d, 22d); útil con datos intradía/escala múltiple. ([ScienceDirect][41], [SSRN][42])

## Mirada unificadora (heurística rápida)

Todos estos modelos implementan, con distinta **memoria** y **simetría**, la misma idea:

$$
\varepsilon_t=\sigma_t z_t,\quad 
\mathbb{E}[\varepsilon_t\mid\mathcal{F}_{t-1}]=0,\quad 
\underbrace{\sigma_t^2}_{\text{“estado latente”}}=\mathcal{G}\big(\varepsilon_{t-1},\sigma_{t-1};\theta\big).
$$

La *elección* de $\mathcal{G}$ codifica un postulado epistemológico sobre cómo el “impacto de la noticia” se propaga: geométrico (GARCH), log-lineal y asimétrico (EGARCH), umbral (GJR/TARCH), potencia y sesgo (APARCH), fraccional/hiperbólico (FIGARCH/FIEGARCH/HYGARCH), multiescala (HARCH), componente (CGARCH), macro-vínculo (MIDAS/Spline), o co-movimiento dinámico (DCC/BEKK). En términos operativos, la decisión es un problema de *matching* entre **estilizados empíricos** (colas, asimetría, persistencia, co-movimientos) y **restricciones estructurales** (positividad, parsimonia, interpretabilidad, cerradura bajo Q).

> *Regla de oro* para elegir: si los *news-impact curves* son asimétricos ⇒ EGARCH/GJR/APARCH; si hay **memoria larga** ⇒ (FI/HY)EGARCH/FIGARCH; si hay “dos escalas” ⇒ CGARCH/MIDAS/Realized; si requieres **valoración** cerrada ⇒ Heston–Nandi; si es cartera multiactivo ⇒ DCC/BEKK/CCC.

¡Vamos a ampliar el mapa con extensiones realmente usadas en práctica y literatura! Para cada una te doy (año), ecuación canónica (o forma típica) y función. Mantengo notación estándar: $r_t=\mu_t+\varepsilon_t$, $\varepsilon_t=\sigma_t z_t$, $z_t$ i.i.d., $L$ lag, y $u_t$ residuos estandarizados.

### **FIAPARCH**

FIPARCH — Fractionally Integrated APARCH (Tse, 1998).
Forma (versión $1,d,1$, esquemática):

$$
(1-\beta L)\,\sigma_t^{\delta} \;=\; \omega \;+\; \bigl[1-(1-\phi L)(1-L)^d\bigr]\,(|\varepsilon_t|-\gamma \varepsilon_t)^{\delta},
\quad 0\le d<1.
$$

Función: combina asimetría (potencia $\delta$, palanca $\gamma$) con memoria larga (fraccionaria $d$). Excelente para colas pesadas con persistencia lenta. ([public.econ.duke.edu][1], [Queen's University Belfast][2], [ink.library.smu.edu.sg][3])

### **ST-GARCH**

ST-GARCH — Smooth-Transition GARCH (González-Rivera, 1998; antecedente Hagerud, 1997).

$$
\sigma_t^2=\omega+\alpha\varepsilon_{t-1}^2+\beta\sigma_{t-1}^2
+\bigl[\alpha \varepsilon_{t-1}^2+\beta \sigma_{t-1}^2\bigr]\;G(s_{t-1};\gamma,c),
$$

con $G$ logística.
Función: cambia suavemente de “régimen calmo” a “tenso” sin saltos discretos. ([ResearchGate][4])

3. MS-GARCH / SWARCH — Markov-Switching (Hamilton–Susmel, 1994; Cai, 1994).

$$
\sigma_t^2=\omega_{S_t}+\alpha_{S_t}\varepsilon_{t-1}^2+\beta_{S_t}\sigma_{t-1}^2,\quad S_t\sim\text{MC}.
$$

Función: regímenes (bear/bull, crisis) con parámetros por estado; captura rupturas estructurales. ([gasmodel.com][5], [public.econ.duke.edu][6])

4. QGARCH — Quadratic/Linear-in-$\varepsilon$ (Sentana, 1995).

$$
\sigma_t^2=\omega+\alpha\varepsilon_{t-1}^2+\gamma\varepsilon_{t-1}+\beta\sigma_{t-1}^2.
$$

Función: palanca a través de un término lineal en el residuo (asimetría “suave”). ([Wiley Online Library][7])

5. log-GARCH (Geweke, 1986; Pantula, 1986; Milhøj, 1987).

$$
\log\sigma_t^2=\omega+\sum_i\alpha_i\,g(\varepsilon_{t-i})+\sum_j\beta_j\log\sigma_{t-j}^2,
$$

p. ej. $g(x)=|x|$ o $x^2-1$.
Función: positividad automática vía log-varianza; buena para colas y escalamiento. ([Wiley Online Library][8])

6. Beta-t-EGARCH / DCS-EGARCH (Harvey & col., 2008–2016).

$$
\log\sigma_t^2=\omega+\beta\,\log\sigma_{t-1}^2+\alpha\,s_t,
$$

$s_t$ = score condicional (t-Student/GT).
Función: “score-driven”; robustece a outliers, estabilidad informacional superior a t-GARCH clásico. ([IDEAS/RePEc][9], [Erasmus University Rotterdam][10], [econ.cam.ac.uk][11])

7. GAS / DCS-Volatility (Creal, Koopman, Lucas, 2012–2013).

$$
\theta_{t+1}=\omega+A\,s_t+B\,\theta_t,\quad s_t\propto\nabla_{\theta}\log f(y_t\mid\theta_t).
$$

Función: marco unificado “impulsado por el score” para parámetros dinámicos (incluida $\log\sigma_t^2$). ([tinbergen.nl][12], [ResearchGate][13], [Journal of Statistical Software][14])

8. HEAVY / Realised-HEAVY (Shephard & Sheppard, 2009–2011).

$$
V_t=\omega+\alpha\,X_{t-1}+\beta\,V_{t-1},
$$

con $X_t$ medidor HF de volatilidad (RV/rango).
Función: usa medidas intradía para “empujar” la varianza diaria; gran reactividad a shocks. ([IDEAS/RePEc][15], [public.econ.duke.edu][16])

9. O-GARCH / GO-GARCH (Alexander, 2001–2002; van der Weide, 2002).
   Descomposición $H_t=A D_t A^\top$ con $D_t$ diagonal y cada diagonal $\sim$ GARCH univar.
   Función: multivariante parsimonioso mediante factores ortogonales / generalizados; escala a muchas series. ([Wiley Online Library][17])

10. ADCC / FDCC — DCC asimétrico y flexible (Cappiello–Engle–Sheppard, 2003/2006; Billio–Caporin–Gobbo, 2004/2006).

$$
Q_t=(1-a-b)S+a\,u_{t-1}u_{t-1}^\top+b\,Q_{t-1}+g\,n_{t-1}n_{t-1}^\top,
$$

($n_t=\min(u_t,0)$); FDCC permite parámetros por grupo.
Función: correlaciones dinámicas con efecto “malas noticias elevan correlaciones”. ([SSRN][18], [Oxford Academic][19], [ResearchGate][20])

11. CARR — Conditional Autoregressive Range (Chou, 2005).

$$
R_t=\psi_t \varepsilon_t,\qquad \psi_t=\omega+\alpha R_{t-1}+\beta\psi_{t-1}.
$$

Función: modela el rango High–Low (proxy eficiente de $\sigma$); frecuentemente supera GARCH en VaR/vol out-of-sample. ([JSTOR][21], [ir.lib.nycu.edu.tw][22])

12. MEM — Multiplicative Error Model (Engle, 2002+).

$$
x_t=\mu_t\varepsilon_t,\qquad \mu_t=\omega+\alpha x_{t-1}+\beta\mu_{t-1},
$$

para series positivas (RV, vol implícita, duraciones).
Función: “GARCH-like” para procesos no-negativos (y su versión vectorial). ([NBER][23], [ScienceDirect][24])

> 


13. Copula-GARCH / DCC-Copula (Patton, 2001/2006; Jondeau–Rockinger, 2006).
    Marginales GARCH + cópula $C_t$ con parámetro dinámico, p. ej.

$$
\rho_t=\Lambda\bigl(\omega+\alpha\,h(u_{t-1})+\beta\,\rho_{t-1}\bigr).
$$

Función: dependencia flexible (colas/asimetrías) separada de las marginales; útil en riesgo de portafolio. ([Pages personnelles][25], [public.econ.duke.edu][26])

> 


14. HARCH — Heterogeneous ARCH (Müller et al., 1997).

$$
\sigma_t^2=\omega+\sum_{k} \alpha_k \Bigl(\sum_{i=1}^{m_k} r_{t-i}\Bigr)^{\!2},
$$

suma de “volatilidades parciales” multiescala.
Función: incorpora horizontes heterogéneos (intradía–semanal); antecede a HAR-RV. ([ScienceDirect][27], [public.econ.duke.edu][1])

15. Distribuciones sesgadas en GARCH (Hansen, 1994; extensiones SGT/GT).
    Mis especificaciones típicas dejan la dinámica GARCH igual pero usan $z_t\sim$ t-sesgada (o SGT).
    Función: captura asimetría y leptocurtosis sin alterar la ecuación de varianza. ([users.ssc.wisc.edu][28], [CRAN][29])

16. Extensiones “long-memory” afines: FIGARCH (Baillie et al., 1996) y FIEGARCH (Bollerslev & Mikkelsen, 1996) ya las mencionamos; FIAPARCH (arriba) y HYGARCH (Davidson, 2004) completan el abanico. (Referencia panorámica). ([SSRN][30])

17. Modelos “realized-vol” afines a GARCH: Realized-GARCH (Hansen–Huang–Shek, 2012) y su análogo HEAVY (arriba) integran medidas de alta frecuencia directamente en la dinámica condicional. ([IDEAS/RePEc][15])

18. Otras variantes útiles en práctica:

    • TVP-/ANN-GARCH (parámetros que varían en el tiempo o aprendidos por ANN). ([Arnaud Dufays][32])
