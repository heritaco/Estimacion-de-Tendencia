¡Va! Pensemos “pronóstico óptimo” como la **proyección de mínima pérdida cuadrática**:

$$
\hat y_{t+h\mid t}\;=\;\mathbb{E}\!\left[y_{t+h}\mid\mathcal F_t\right],
$$

y, en modelos lineales‐gaussianos, coincide con el **mejor predictor lineal** (Wiener–Kolmogórov) y con el filtro de **Kalman** en formulación estado–espacio. Debajo tienes un **mapa cronológico–funcional** (año, **fórmula nuclear**, **para qué sirve**) centrado en la **media condicional**.

# Genealogía de modelos para pronóstico de la media

**(1) Wold (1938) / Wiener–Kolmogórov (1941–48)** — *óptimo lineal*

$$
y_t=\sum_{k=0}^{\infty}\psi_k \varepsilon_{t-k},\qquad
\hat y_{t+h\mid t}=\sum_{k=h}^{\infty}\psi_k\,\varepsilon_{t+h-k}.
$$

Base teórica: descomposición y proyección óptima vía autocovarianzas (Yule–Walker).

**(2) AR(p) (1927–31)** — *autorregresión*

$$
\phi(B)\,y_t=c+\varepsilon_t,\quad \phi(B)=1-\phi_1B-\cdots-\phi_pB^p.
$$

Pronóstico recursivo lineal; excelente si la ACF decae exponencialmente.

**(3) MA(q) / ARMA(p,q) (1970, Box–Jenkins)** — *ruido con memoria corta + choques*

$$
\phi(B)y_t=c+\theta(B)\varepsilon_t.
$$

Pronóstico usa $\hat\varepsilon_{t-j}$ y asume $\varepsilon_{t+h}=0$ para $h>0$.

**(4) ARIMA(p,d,q) (1970)** — *no estacionarias integradas*

$$
\phi(B)(1-B)^d y_t=c+\theta(B)\varepsilon_t.
$$

Pronóstico sobre la serie diferenciada; re‐integra para niveles.

**(5) SARIMA (1976)** — *estacionalidad explícita*

$$
\Phi(B^s)\phi(B)(1-B)^d(1-B^s)^D y_t=c+\Theta(B^s)\theta(B)\varepsilon_t.
$$

Capta ciclos calendario (mensual, semanal, etc.).

**(6) Transfer Function / ARIMAX (1970–76)** — *regresores exógenos filtrados*

$$
y_t=\frac{v(B)}{\delta(B)}x_t + n_t,\quad n_t\sim\text{ARIMA}.
$$

Útil cuando causas observables anticipan $y_t$.

**(7) ARFIMA / FARIMA (1980–81)** — *memoria larga en la media*

$$
\phi(B)(1-B)^d y_t=\theta(B)\varepsilon_t,\quad d\in(0,1).
$$

Pronóstico con decaimiento hiperbólico (persistencia lenta).

**(8) Suavizamiento exponencial simple (Brown, 1956)** — *nivel local*

$$
\ell_t=\alpha y_t+(1-\alpha)\ell_{t-1},\quad \hat y_{t+h\mid t}=\ell_t.
$$

Óptimo para RW+ruido; base de ETS.

**(9) Holt (1957) — tendencia**

$$
\ell_t=\alpha y_t+(1-\alpha)(\ell_{t-1}+b_{t-1}),\quad
b_t=\beta(\ell_t-\ell_{t-1})+(1-\beta)b_{t-1},\\
\hat y_{t+h\mid t}=\ell_t+h\,b_t.
$$

Tendencia lineal con amortiguación opcional.

**(10) Holt–Winters (1960)** — *tendencia + estacionalidad (A/M)*
Actualizaciones de $\ell_t,b_t,s_t$ (aditiva o multiplicativa);
$\hat y_{t+h\mid t}=\ell_t+h\,b_t+s_{t+h-mk}$.
Clásico industrial para series con temporadas.

**(11) ETS (Error–Trend–Seasonal) estado–espacio (2002–2008)**
Forma general: $y_t=\ell_{t-1}+b_{t-1}+s_{t-1}+\varepsilon_t$ (o variantes M), con actualizaciones recursivas.
Óptimo (MMSE) bajo supuestos de innovación aditiva/multiplicativa.

**(12) Modelos estructurales / UCM (Harvey, 1989)** — *componentes latentes*

$$
y_t=\mu_t+\gamma_t+c_t+\varepsilon_t,\quad
\mu_t=\mu_{t-1}+\eta_t,\;\; \gamma_t\text{ (estacional)},\;\; c_t\text{ (ciclo)}.
$$

Kalman: pronóstico y *smoothing* de componentes.

**(13) Local linear trend / Random walk plus noise (1987–89)**

$$
y_t=\mu_t+\varepsilon_t,\;\;
\mu_t=\mu_{t-1}+ \beta_{t-1}+\eta_t,\;\;
\beta_t=\beta_{t-1}+\zeta_t.
$$

Excelente en macro con quiebres suaves.

**(14) Theta method (2000)** — *descomposición de curvatura*
Combinación de proyecciones “theta”; ganador M3; sorprendentemente robusto.

**(15) BATS/TBATS (2011)** — *Box–Cox, ARMA errors, Trend, multiple Seasonals*
Modelo estado–espacio con estacionalidades múltiples (p.ej., intradía + semanal).

**(16) VAR(p) (Sims, 1980)** — *multivariante, media conjunta*

$$
y_t=c+A_1y_{t-1}+\cdots+A_p y_{t-p}+\varepsilon_t.
$$

Pronósticos vectoriales; *impulse responses*. **VECM** (1988) si hay cointegración.

**(17) ARDL / DL (1998)** — *efecto dinámico de X en y*

$$
y_t=\sum_{i=0}^p \alpha_i y_{t-i}+\sum_{j=0}^q \beta_j x_{t-j}+\varepsilon_t.
$$

Buena trazabilidad de elasticidades de corto y largo plazo.

**(18) BSTS / DLM Bayesianos (1990s–2014)** — *tendencia, estacionalidad, regresores, cambios*
Ecuaciones lineales gaussianas con priors; pronósticos completos (distribucionales) y *model averaging*.

**(19) Gaussian Processes (2006+)** — *no paramétrico, kernelizado*

$$
y\sim \mathcal{GP}(m(\cdot),k(\cdot,\cdot)).
$$

Pronóstico = posterior gaussiano; flexibilidad en periodicidad/estructura con *spectral kernels*.

**(20) ML clásicos para TS (2010s)** — RF/GBM/XGBoost/LightGBM

$$
\hat y_{t} = f(y_{t-1:t-L},\;x_{t},\;\text{calendario};\;\theta).
$$

No “óptimos” por teoría cerrada, pero muy competitivos con *features* y *cross‐validation* temporal.

**(21) Redes neuronales específicas**

* **RNN/LSTM/GRU** (1997–2014): memoria larga/no lineal.
* **TCN** (2018): convoluciones causales dilatadas.
* **Transformers** (2017–2023, TFT/Informer/Autoformer): atención causal y multi‐horizonte.
  Potentes con exógenas y múltiples estacionalidades; requieren regularización cuidadosa.

**(22) Prophet (2017)** — *tendencia + estacionalidades + festivos (robusto)*
Modelo aditivo con *changepoints*; prioriza interpretabilidad/robustez industrial.

**(23) Combinación de pronósticos (Clemen, 1989; Makridakis, 2000s)**

$$
\hat y^{(comb)}=\sum_k w_k \hat y^{(k)},\quad \sum w_k=1.
$$

Teorema “diversity–bonus”: mezcla reduce varianza si errores son imperfectamente correlacionados.

## Apuntes operativos (ARIMA y “óptimo”)

* En **ARMA/ARIMA**, bajo gaussianidad, el predictor óptimo (MMSE) es lineal y se obtiene por **innovations algorithm** o recursión Box–Jenkins:

  $$
  \hat y_{t+1\mid t}=\mu+\sum_{i=1}^{p}\phi_i y_{t+1-i}+\sum_{j=1}^{q}\theta_j \hat\varepsilon_{t+1-j},\quad
  \hat\varepsilon_{t+h}=0\ (h>0).
  $$
* En **estado–espacio** (ETS/UCM/ARIMA como SSF), **Kalman** entrega $\hat y_{t+h\mid t}=\mathbb E[y_{t+h}\mid\mathcal F_t]$ y su varianza $\mathbb V$ óptimas.
* **Con exógenas** (ARIMAX/TF/VARX/ARDL), el óptimo depende también del pronóstico de $x_t$: o bien se *condiciona* en escenarios de $x$, o se pronostica $x$ con su propio modelo.

## Micro–guía de elección (media)

* Tendencia + estacionalidad “limpia”: **ETS / Holt–Winters / Theta / TBATS**.
* Estructura AR clara, ACF/PACF “clásica”: **ARIMA/SARIMA**.
* Exógenas con retardos: **ARIMAX / Transfer Function / ARDL**.
* Cointegración multivariante: **VECM/VAR**.
* Cambios de régimen suaves/abruptos: **UCM/BSTS**.
* Periodicidades múltiples irreducibles + exógenas ricas: **TBATS** o **Transformers/TCN**.
* Benchmark siempre: **naive** ($y_{t}$) y **seasonal naive** ($y_{t-s}$).
